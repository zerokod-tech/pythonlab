```python
import pandas as pd
df = pd.read_csv("modified_titanic_data.csv")
df.head()
```

```python
# df = df.drop(columns="Unnamed: 13")
# or 
df.drop(columns="Unnamed: 13", inplace = True)
df.head()
```

```python
#remove duplicated //Keep=false > removing all duplicated: (all true)
df[df.duplicated(keep=False)]
```

```python
#remove duplicated //first : Mark duplicates as True except for the first occurrence.
df[df.duplicated(keep='first')]
```

```python
#remove duplicated //last : Mark duplicates as True except for the last occurrence.
df[df.duplicated(keep='last')]
```
